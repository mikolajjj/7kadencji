```python
import pandas as pd
import json
import seaborn
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("all.csv")
```

```python
df.head()
```

```python
len(df.klub.unique())
```

```python
def clean_name(klub):
    return klub.replace("Koło", "").replace("Klub", "").replace("Parlamentarne", "").replace("Parlamentarny", "").replace("Poselskie", "").strip()
```

```python
df.klub = df.klub.apply(clean_name)
```

```python
len(df.klub.value_counts())
```

```python
df.klub.value_counts().plot(kind='barh', figsize=(6,12))
```

```python
for kadencja in ["I", "II", "III", "IV", "V", "VI", "VII", "VIII"]:
    for etap in ["start", "end"]:
        df[(df.kadencja == kadencja) & (df.etap == etap)].klub.value_counts().plot(kind='barh', figsize=(6,12))
        plt.title(kadencja + " " + etap)
        plt.show()
```
